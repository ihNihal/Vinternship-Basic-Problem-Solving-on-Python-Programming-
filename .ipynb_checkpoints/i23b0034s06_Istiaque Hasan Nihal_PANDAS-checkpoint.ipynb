Practice Problem 2.1

 

Mr. Box generated a number sequential list including the range 1001 to 2000. Generate the data series using pandas. Then write a Pandas program to find the positions of numbers that are multiples of 5 of a given series.

```python
import pandas as pd

#Generate number sequential list
data_series = pd.Series(range(1001,2000))

# Finding the position of the numbers that are multiple of 5
multiple_5 = data_series[data_series % 5 == 0].index
for i in multiple_5:
    print(multiple_5)
```


Practice problem 2.2

Write a Pandas program to get the day of month, day of year, week number and day of week from a given series of date strings.y

```python
import pandas as pd

# Create the given series of date strings
date_strings = pd.Series([
    '01 Jan 2015',
    '10-02-2016',
    '20180307',
    '2014/05/06',
    '2019-04-12',
    '2019-04-06T11:20'

  
])
date_series = pd.to_datetime(date_strings, format = 'mixed')

day_of_month = date_series.dt.day.tolist()
day_of_year = date_series.dt.dayofyear.tolist()
week_number = date_series.dt.isocalendar().week.tolist()
day_of_week = date_series.dt.day_name().tolist()

print("Original Series:")
print(date_strings)
print("\nDay of month:")
print(day_of_month)
print("\nDay of year:")
print(day_of_year)
print("\nWeek number:")
print(week_number)
print("\nDay of week:")
print(day_of_week)
```

Practice problem 2.4

Write a Pandas program to join the two given dataframes along rows and merge with another dataframe along the common column id.

```python
student_data1 = pd.DataFrame({
    'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'],
    'marks': [200, 210, 190, 222, 199]
})

student_data2 = pd.DataFrame({
    'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
    'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'],
    'marks': [201, 200, 198, 219, 201]
})

exam_data = pd.DataFrame({
    'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
    'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]
})

#Selecting two dataframe for horizontal merge
dataframe_hor = [student_data1, student_data2]

#Horizontal merge
result_data = pd.concat(dataframe_hor)
print(result_data)

final_data = pd.merge(result_data, exam_data, on= 'student_id', how= 'inner')
final_data
```
