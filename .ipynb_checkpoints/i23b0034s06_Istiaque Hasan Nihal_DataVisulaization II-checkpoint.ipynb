
Practice problem 6.4

Find the Quarter-wise average sales price and plot it into a pie chart.

```python
# Merging fact table with time_dim
add_fact_with_time = pd.merge(fact_table, time_dim, on = 'time_key')

#Groupby quarter with average of total price
add_fact_with_time_g = add_fact_with_time.groupby(['quarter']).agg({'total_price':'mean'}).reset_index()

plt.pie(labels=add_fact_with_time_g.quarter, x=add_fact_with_time_g.total_price)

plt.show()
```


Practice Problem 6.7

Compare the total sales price in year-wise quarterly between Dhaka and Chittagong, and plot it into a Scatter Chart.

```python
add_fact_with_time = pd.merge(fact_table, time_dim, on = "time_key" )
add_fact_wtih_time_customer = pd.merge(add_fact_with_time, customer_dim, on = "coustomer_key")


dhaka_div = add_fact_wtih_time_customer.query("division == 'Dhaka'")
dhaka_div_g = dhaka_div.groupby('quarter').aggregate({'total_price': 'sum'}).reset_index()

chittagong_div = add_fact_wtih_time_customer.query("division == 'Chittagong'")
chittagong_div_g = chittagong_div.groupby('quarter').aggregate({'total_price': 'sum'}).reset_index()

plt.plot(dhaka_div_g['quarter'], dhaka_div_g['total_price'], marker='o')

plt.plot(chittagong_div_g['quarter'], chittagong_div_g['total_price'], marker='s')


plt.xlabel('Quarter')

plt.ylabel('Total Sales Price)')


plt.title("Year-wise Quarterly total sales price in Dhaka and Chittagong division")

plt.legend(['Dhaka', 'Chittagong'])
```

Practice Problem 6.8

Break down the total sales of items geographically (division-wise).

Plot division-wise total sales price into SwarmChart.
Plot item-wise total sales price into SwarmChart.

```python
fact_table_with_store_dim = pd.merge(fact_table, store_dim, on = "store_key")
fact_table_with_store_dim_g = fact_table_with_store_dim.groupby("division").aggregate({"total_price":"sum"}).reset_index()
plt.figure(figsize= (12,6))
sns.swarmplot(x="division", y="total_price", data =fact_table_with_store_dim_g )
plt.title('Division-wise Total Sales Price')
plt.xlabel('Division')
plt.ylabel('Total Sales Price')
plt.grid(True)

# Show the plot
plt.show()

```

```python
#Merge Item dim with fact table
fact_with_item = pd.merge(fact_table, item_dim, on = "item_key")
fact_with_item_g = fact_with_item.groupby("item_name").aggregate({"total_price": "sum"}).reset_index()
plt.figure(figsize = (20,6))
sns.swarmplot(x="item_name", y="total_price", data =fact_with_item_g )
plt.title('Item-wise Total Sales Price')
plt.xlabel('Item')
plt.ylabel('Total Sales Price')
plt.grid(True)

# Show the plot
plt.show()
```

Practice Problem 6.9

Break down year-wise the total sales geographically (division-wise) and plot it into Stripplot.

```python
add_fact_with_time = pd.merge(fact_table, time_dim, on = "time_key" )
add_fact_wtih_time_customer = pd.merge(add_fact_with_time, customer_dim, on = "coustomer_key")
year_wise = add_fact_wtih_time_customer.groupby(["division", "year"]).aggregate({"total_price":"sum"}).reset_index()

sns.stripplot(y = year_wise['total_price'], x = year_wise["division"])
```

What was the distribution of Ages for passengers in different classes?


```python
df = sns.load_dataset("titanic")
sns.set_style('white')

sns.kdeplot(data = df,x='age',hue = "class")



plt.title('Distribution of Ages for Passengers in Different Classes')
plt.xlabel('Age')
plt.ylabel('Density')
sns.despine(left=True, bottom=True)

plt.show()
```
