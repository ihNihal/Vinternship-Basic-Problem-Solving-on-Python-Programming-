```python
#Importing Libraries 

import seaborn as sns
import psycopg2
import pandas as pd
import psycopg2.extras
```

```python
import matplotlib.pyplot as plt

#creating a class so we can use objects
class PostgresConnection(object):
    #right after calling this class __init__() function will execute first
    def __init__(self):
        self.connection = psycopg2.connect(database = "ecomdb",
                                           user = "postgres",
                                           password = "admin",
                                           host = "127.0.0.1",
                                           port = "5432")
        print("\nConnected!")
    #getConnection() function will call __init__() and connect
    def getConnection(self):
            print("Connection to DB established")
            return self.connection
```

#### Q1: Find the division wise total_sale_price joining fact table and respective dimension table

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
select_stmt = """  
SELECT s.division, SUM(t.total_price)
FROM ecom_star_schema.fact_table t
JOIN ecom_star_schema.store_dim s on s.store_key = t.store_key
GROUP BY CUBE(s.division)
order by s.division

"""
cur.execute(select_stmt)
records = cur.fetchall()
```

```python
df_q1 = pd.DataFrame(records, columns = ['division', 'sales'])
df_q1
```

#### Q6: For each store(item supplier), what are the top three products offered that are most often purchased?

    S1-> item1 -> quantity_sales

    S1-> item2 -> quantity_sales

    S1-> item3 -> quantity_sales

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
select_stmt = """  
SELECT s.store_key, i.item_name,  SUM(f.quantity)
FROM ecom_star_schema.fact_table f
JOIN ecom_star_schema.item_dim i on i.item_key = f.item_key
JOIN ecom_star_schema.store_dim s on s.store_key = f.store_key
GROUP BY CUBE(s.store_key, i.item_name, f.quantity)
order by s.store_key, SUM(f.quantity) DESC

"""
cur.execute(select_stmt)
records = cur.fetchall()
df_q6 = pd.DataFrame(records, columns = ['Store_key','Item_name' ,'Quantity'])
df_q6=df_q6.dropna()
df_q6
```

```python
df_q6 = df_q6[0:21]
plt.figure(figsize=(15,10))
plt.title("Item Wise sales", fontsize = 20)
plt.xlabel("Item Name", fontsize = 20)
plt.ylabel("Quantity", fontsize = 20)
sns.scatterplot(data = df_q6, x= 'Quantity', y = 'Item_name', hue= 'Store_key', palette = 'tab10', s=200, style= 'Item_name' )
sns.set(font_scale = 1)
```

#### Q7: What products have been sold through card or mobile since X days?

    input: X = 400 days [use 400 or more days]

    output: [item1, item2, item3, ...........]

```python
x = input('Number of days')
```

```python
# Through Card
con = PostgresConnection().getConnection()
cur = con.cursor()
s1 = """  
SELECT i.item_name, ti.date, t.trans_type
FROM ecom_star_schema.fact_table f
JOIN ecom_star_schema.item_dim i on i.item_key = f.item_key
JOIN ecom_star_schema.trans_dim t on t.payment_key = f.payment_key
JOIN ecom_star_schema.time_dim ti on ti.time_key = f.time_key
WHERE ti.date > (CURRENT_DATE - INTERVAL '"""
s2 = str(x) + " days') AND t.trans_type = 'card'"
select_stmt = s1 + s2
cur.execute(select_stmt)
records = cur.fetchall()
df_q7 = pd.DataFrame(list(records), columns=['Item_name', 'Dates', 'Trans_type'])
df_q7 = df_q7.dropna()
df_q7
```

#### Q10: What are the average sales of products sales per store monthly?

     S1->M1-> avg_sales

    s1-> M2 -> avg_sales

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
select_stmt = """  
SELECT  c.store_key,i.month,  AVG(f.quantity)
FROM ecom_star_schema.fact_table f
JOIN ecom_star_schema.time_dim i on i.time_key = f.time_key
JOIN ecom_star_schema.store_dim c on c.store_key = f.store_key
GROUP BY CUBE(c.store_key, i.month, f.quantity)
order by i.month,c.store_key, AVG(f.quantity) ASC

"""
cur.execute(select_stmt)
records = cur.fetchall()
df_q10 = pd.DataFrame(records, columns = ['Store' ,'Month','Quantity'])
df_q10=df_q10.dropna()
df_q10
```

# 1. store and time dimensional financial analytics.

## Which store had the highest total sales in the last quarter of the year?

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
select_stmt = """  
SELECT  c.store_key,i.quarter,  MAX(f.total_price)
FROM ecom_star_schema.fact_table f
JOIN ecom_star_schema.time_dim i on i.time_key = f.time_key
JOIN ecom_star_schema.store_dim c on c.store_key = f.store_key
WHERE i.quarter = 'Q4'
GROUP BY CUBE(c.store_key, i.quarter, f.total_price)
order by i.quarter, c.store_key, MAX(f.total_price) DESC

"""
cur.execute(select_stmt)
records = cur.fetchall()
df = pd.DataFrame(records, columns = ['Store' ,'Quarter','Sales'])
df=df.dropna()
df
```

```python
plt.figure(figsize=[10,15])
sns.barplot(y=df['Store'], x= df['Sales'], hue = df['Quarter'], legend =True)
plt.xlabel("Store")
plt.ylabel("Max sales per store on last quarter")
```

## What is the trend of monthly revenue for each store over the past year?

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
select_stmt = """  
SELECT  c.store_key,i.month,  SUM(f.total_price)
FROM ecom_star_schema.fact_table f
JOIN ecom_star_schema.time_dim i on i.time_key = f.time_key
JOIN ecom_star_schema.store_dim c on c.store_key = f.store_key
GROUP BY CUBE(c.store_key, i.month, f.total_price)
order by i.month, c.store_key, SUM(f.total_price) DESC

"""
cur.execute(select_stmt)
records = cur.fetchall()
df = pd.DataFrame(records, columns = ['Store' ,'Month','Sales'])
df=df.dropna()
df
```

```python
plt.figure(figsize=[10,15])
sns.lineplot(y=df['Sales'], x= df['Month'], hue = df['Store'], legend =True)
plt.xlabel("Month")
plt.ylabel("Monthly sales per store")
```

## What is the revenue contribution of each item category to the total sales in the past year?

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
select_stmt = """  
SELECT  c.item_name,i.year,  SUM(f.total_price)
FROM ecom_star_schema.fact_table f
JOIN ecom_star_schema.time_dim i on i.time_key = f.time_key
JOIN ecom_star_schema.item_dim c on c.item_key = f.item_key
WHERE i.year = '2019'
GROUP BY CUBE(c.item_name, i.year, f.total_price)
order by i.year ,SUM(f.total_price) DESC

"""
cur.execute(select_stmt)
records = cur.fetchall()
df = pd.DataFrame(records, columns = ['Item' ,'month','Sales'])
df=df.dropna()
df
```

# 5. item and time dimensional inventory analytics

## Which item has the highest inventory turnover rate in the past quarter?

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
select_stmt = """  
SELECT  c.item_name, i.quarter, AVG(f.total_price/f.quantity)
FROM ecom_star_schema.fact_table f
JOIN ecom_star_schema.time_dim i on i.time_key = f.time_key
JOIN ecom_star_schema.item_dim c on c.item_key = f.item_key
GROUP BY CUBE( i.quarter, c.item_name, f.total_price )
order by i.quarter , AVG(f.total_price/f.quantity) DESC

"""
cur.execute(select_stmt)
records = cur.fetchall()
df = pd.DataFrame(records, columns = ['Item' ,'Quarter','Sales'])
df=df.dropna()
df
```

```python
df = df[0:50]
plt.figure(figsize=[15,15])
sns.lineplot(y=df['Sales'], x= df['Item'], hue = df['Quarter'], legend =True)
plt.xlabel("Item")
plt.ylabel("Sales")
```

## 2. What is the trend of stock levels for each item over the past year?

```python
con = PostgresConnection().getConnection()
cur = con.cursor()
select_stmt = """  
SELECT c.item_name, c.stock_quantity, i.year
FROM ecom_star_schema.fact_table f
JOIN ecom_star_schema.time_dim i on i.time_key = f.time_key
JOIN ecom_star_schema.item_dim c on c.item_key = f.item_key
WHERE i.year = '2022'
GROUP BY CUBE( c.item_name, c.stock_quantity, i.year )
order by  c.stock_quantity, i.year ASC

"""
cur.execute(select_stmt)
records = cur.fetchall()
df = pd.DataFrame(records, columns = ['Item' ,'Quantity','Year'])
df=df.dropna()
df
```

```python
df = df[0:50]
plt.figure(figsize=[15,15])
sns.barplot(x=df['Quantity'], y= df['Item'], hue = df['Year'], legend =True)
plt.xlabel("Item")
plt.ylabel("Sales")
```
