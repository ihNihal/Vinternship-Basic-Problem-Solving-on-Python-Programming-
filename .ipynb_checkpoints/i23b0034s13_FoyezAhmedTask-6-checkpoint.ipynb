#### Practice Problem 6.6
Compare the total sales price in year-wise quarterly between Dhaka and Chittagong, and plot it into a Scatter Chart.

```python
add_time_with_fact = pd.merge(fact_table, time_dim, on = 'time_key')
add_customer_with_fact = pd.merge(add_time_with_fact, customer_dim, on = 'coustomer_key')

dhaka_div = add_customer_with_fact.query("division=='Dhaka'")
dhaka_div_g = dhaka_div.groupby(['quarter']).agg({'total_price':'sum'}).reset_index()

ctg_div = add_customer_with_fact.query("division=='Chittagong'")
ctg_div_g = ctg_div.groupby(['quarter']).agg({'total_price':'sum'}).reset_index()

plt.plot(dhaka_div_g['quarter'], dhaka_div_g['total_price'])
plt.plot(ctg_div_g['quarter'], ctg_div_g['total_price'])


plt.xlabel('Quarter')
plt.ylabel('Total Sales Price)')


plt.title("Year-wise Quarterly total sales price in Dhaka and Chittagong division")
plt.legend(['Dhaka', 'Chittagong'])
```

#### Practice Problem 6.7
Break down the total sales of items geographically (division-wise).
1. 
Plot division-wise total sales price into SwarmChar

t2. .
Plot item-wise total sales price into SwarmChart.

```python
add_item_with_fact = pd.merge(fact_table, item_dim, on = 'item_key')
add_customer_with_fact = pd.merge(add_item_with_fact, customer_dim, on = 'coustomer_key')
div_sales = add_customer_with_fact.groupby(['division']).agg({'total_price':'sum'}).reset_index()
plt.figure(figsize= (12,6))
sns.swarmplot(x = div_sales['division'], y = div_sales['total_price'])
plt.title('Division-wise Total Sales Price')
plt.xlabel('Division')
plt.ylabel('Total Sales Price')
plt.grid(True)
```

```python
div_sales = add_customer_with_fact.groupby(['item_name']).agg({'total_price':'sum'}).reset_index()
plt.figure(figsize= (20,6))
sns.swarmplot(x = div_sales['item_name'], y = div_sales['total_price'])
plt.grid(True)

plt.title('Item-wise Total Sales Price')
plt.xlabel('Item')
plt.ylabel('Total Sales Price')
```

#### Practice Problem 6.8
Break down year-wise the total sales geographically (division-wise) and plot it into Stripplot.

```python
add_time_with_fact = pd.merge(fact_table, time_dim, on = 'time_key')
add_customer_with_fact = pd.merge( add_time_with_fact, customer_dim, on = 'coustomer_key')
div_sales = add_customer_with_fact.groupby(['division','year']).agg({'total_price':'sum'}).reset_index()
plt.figure(figsize= (12,6))

sns.stripplot(y = div_sales['total_price'], x = div_sales['year'])

plt.title('Year-wise Total Sales Price')
plt.xlabel('Division')
plt.ylabel('Total Sales Price')
plt.grid(True)
```

#### Practice Problem 6.9
What was the distribution of Ages for passengers in different classes?

```python
df = sns.load_dataset("titanic")

sns.set_style('white')
sns.kdeplot(df,x='age',hue = "class")
plt.title('Distribution of Ages for Passengers in Different Classes')
plt.xlabel('Age')
plt.ylabel('Density')
sns.despine(left=True, bottom=True)
plt.show()
```
