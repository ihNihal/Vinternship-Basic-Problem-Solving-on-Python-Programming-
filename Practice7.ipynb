#### Connect to DB

```python
#creating a class so we can use objects
class PostgresConnection(object):
    #right after calling this class __init__() function will execute first
    def __init__(self):
        self.connection = psycopg2.connect(database = "ecomdb",
                                           user = "postgres",
                                           password = "admin",
                                           host = "127.0.0.1",
                                           port = "5432")
        print("\nConnected!")
    #getConnection() function will call __init__() and connect
    def getConnection(self):
            print("Connection to DB established")
            return self.connection
```

```python
#con object to initiate the connection
con = PostgresConnection().getConnection()
```

#### Inserting Data into tables

#### Trans_dim

```python
try:
    insert_stmt = """INSERT INTO ecom_star_schema.trans_dim(payment_key, trans_type, bank_name) Values (%s, %s, %s)"""

#to open a transaction we have to use cursor() so that in postgresql theres a cursos to run the transaction
    cur = con.cursor()

#since we are transacting data in batches we are using this execute_batch() function
#execute_batch(cursor, insert_statement, table_name.values)
    psycopg2.extras.execute_batch(cur, insert_stmt, trans_dim.values)


    con.commit()
    print("Inserted Successfully!")
    cur.close()
except Exception as e:
    con.rollback()
    print("Error:", e)
finally:
    cur.close()
```
